# MNIST_Celldega

Linked **MNIST viewer (deck.gl)** ↔ **Celldega Clustergram** of Leiden clusters of handwritten digits.

* **Clustergram** — columns are image clusters (~100, Leiden), rows are top-N pixels by sum, values are mean ink intensity inside each cluster. Column color strip = majority digit per cluster. Row value strip = `Center` (radial proximity to the image center, so reordering by `Center` surfaces patterns like *Zero* and *Seven* having low ink at the image center).
* **MNIST viewer** — atlas of all 100 cluster averages by default. Clicking a cluster column zooms in and shows that cluster's example digits. Clicking a pixel row marks the pixel on every average. Clicking a `Majority-digit:` color filters to that digit. Clicking a column-axis dendrogram **averages all selected clusters** into one merged image. Clicking a row-axis dendrogram tints each cluster by its mean ink across the selected pixels.

All linkages are `jsdlink` so the exported HTML works without a Python kernel.

```python
from mnist_celldega import silence_warnings
silence_warnings()

from ipywidgets import HBox, Layout
from mnist_celldega import (
    get_mnist_data, cluster_mnist,
    make_mnist_clustergram, make_mnist_viewer_widget, link_viewer_to_clustergram,
)

ds = get_mnist_data()
clusters = cluster_mnist(ds, mode='all', n_clusters=100, n_top_pixels=500, n_examples=16)
mat, cgm = make_mnist_clustergram(clusters)
viewer = make_mnist_viewer_widget(clusters, width=540, height=600)
link_viewer_to_clustergram(viewer, cgm)

viewer.layout = Layout(width='540px', height='600px')
cgm.layout    = Layout(width='820px', height='720px')
HBox([viewer, cgm])
```
